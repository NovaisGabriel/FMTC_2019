```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from NeuralNetwork import NeuralNetwork as NN
from NeuralNetworkOriginal import NeuralNetwork as NNO
from sklearn import preprocessing
```

### Importing Data

```python
data = pd.read_csv('Churn_Modelling.csv')
data.head()
```

### Cleaning Data

```python
X = data[['CreditScore','Geography','Gender','Age','Tenure','Balance','NumOfProducts','HasCrCard','IsActiveMember','EstimatedSalary']]
y = data[['Exited']]
```

### Preparing

```python
geo = pd.get_dummies(data['Geography'])
gender = pd.get_dummies(data['Gender'])  
X = pd.concat([X,geo],axis=1)
X = pd.concat([X,gender],axis=1)
```

```python
X = X.drop(['Geography','Gender'],axis=1)
X.head()
```

```python
X = X.drop(['Germany','Female'],axis=1) #dummie trap
```

```python
X_norm = X.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(X_norm)
X_norm = pd.DataFrame(x_scaled)
X_norm.head()
```

```python
X_norm = pd.DataFrame(X_norm)
y = pd.DataFrame(y)
```

```python
X_train, X_validate, X_test = np.split(X_norm.sample(frac=1), [int(.6*len(X_norm)), int(.8*len(X_norm))])
y_train, y_validate, y_test = np.split(y.sample(frac=1), [int(.6*len(y)), int(.8*len(y))])
```

```python
print('y_train: '+str(y_train.shape[0])+' rows and '+str(y_train.shape[1])+' columns' )
print('y_validate: '+str(y_validate.shape[0])+' rows and '+str(y_validate.shape[1])+' columns' )
print('y_test: '+str(y_test.shape[0])+' rows and '+str(y_test.shape[1])+' columns' )
print('X_train: '+str(X_train.shape[0])+' rows and '+str(X_train.shape[1])+' columns' )
print('X_validate: '+str(X_validate.shape[0])+' rows and '+str(X_validate.shape[1])+' columns' )
print('X_test: '+str(X_test.shape[0])+' rows and '+str(X_test.shape[1])+' columns' )
```

### Building NN

```python
neural_network = NNO()

print("Beginning Randomly Generated Weights: ")
print(neural_network.synaptic_weights)

training_inputs = np.array(X_train)
training_outputs = np.array(y_train)

validation_inputs = np.array(X_validate)
validation_outputs = np.array(y_validate)

neural_network.train(training_inputs, training_outputs, 10)
```
